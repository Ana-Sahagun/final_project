```python
import pandas as pd

# Ruta de la base de datos 1. PIB pm Demanda (Precios corrientes):
file_path = r"C:\Users\anasa\Desktop\Ana\IRONHACK\Proyecto final\Datos descargados\1. PIB pm Demanda (Precios corrientes).xlsx"

# Leer la base datos de 1. PIB pm Demanda (Precios corrientes):
df = pd.read_excel(file_path)

# Visualizaxion de la tabla completa, queremos ver que coge todas las variables que necesitamos para el analisis:
df
```

```python
# Para transformar la tabla de formato soble entrada a formato tabular usamos el metodo melt():
df=df.melt(id_vars=["variables"], var_name="quarter", value_name="totals")
df['year'] = df['quarter'].str[:4].astype(int)
df['quarter'] = df['quarter'].str[5:].astype(int)
df
```

```python
# Agrupar por Año y calcular la suma de cada variable (M€) para no tener datos en valores trimestrale:
df_agregado= df.groupby(["year", "variables"]).agg({"totals":"sum", "quarter":"size"}).reset_index()
df_agregado
```

```python
# Para calcular la tasa de variación de cada vavriable de un año a otro:
df_agregado['tasa_de_variacion'] = df_agregado.groupby('variables')['totals'].pct_change() * 100
df_agregado
```
